# Task 1: Privacy in API Usage

```python
import requests
import os

API_KEY = os.getenv("WEATHER_API_KEY")

def get_weather(city):
    url = f"https://api.weatherapi.com/data?key={API_KEY}&q={city}"
    response = requests.get(url)
    return response.json()
```

# Task 2: Privacy & Security in File Handling

```python
import hashlib

def store_user(name, email, password):
    hashed_password = hashlib.sha256(password.encode()).hexdigest()
    with open("users.txt", "a") as file:
        file.write(f"{name},{email},{hashed_password}\n")
```

# Task 3: Transparency in Algorithm Design

```python
def is_armstrong(number):
    digits = str(number)
    power = len(digits)
    total = 0

    for digit in digits:
        total += int(digit) ** power

    return total == number
```

## Task 4: Transparency in Algorithm Comparison

Bubble Sort

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
```

Quick Sort

```python
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[0]
    left = [x for x in arr[1:] if x <= pivot]
    right = [x for x in arr[1:] if x > pivot]
    return quick_sort(left) + [pivot] + quick_sort(right)
```

# Task 5: Transparency in AI Recommendations

```python
def recommend_product(user_interest):
    products = {
        "books": "Python Programming Book",
        "fitness": "Yoga Mat",
        "music": "Wireless Headphones"
    }

    if user_interest in products:
        return f"Recommended: {products[user_interest]} because you are interested in {user_interest}."
    else:
        return "No recommendation available."
```
