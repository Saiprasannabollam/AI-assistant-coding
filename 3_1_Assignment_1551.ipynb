Question 1: Zero-Shot Prompting – Palindrome Number Program

```python
def is_palindrome(num):
    return str(num) == str(num)[::-1]
```

Question 2: One-Shot Prompting – Factorial Calculation

```python
def factorial(n):
    if n == 0:
        return 1
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result
```

Question 3: Few-Shot Prompting – Armstrong Number Check

```python
def is_armstrong(num):
    digits = str(num)
    power = len(digits)
    total = sum(int(d)**power for d in digits)
    return total == num
```

Question 4: Context-Managed Prompting (Optimized Number
Classification)

```python
def classify_number(n):
    if not isinstance(n, int):
        return "Invalid input"
    if n <= 1:
        return "Neither Prime nor Composite"

    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return "Composite"
    return "Prime"
```

Question 5: Zero-Shot Prompting – Perfect Number Check

```python
def is_perfect(num):
    total = 0
    for i in range(1, num):
        if num % i == 0:
            total += i
    return total == num
```

Question 6: Few-Shot Prompting – Even or Odd with Validation

```python
def even_or_odd(num):
    if not isinstance(num, int):
        return "Invalid input"
    if num % 2 == 0:
        return "Even"
    else:
        return "Odd"
```
