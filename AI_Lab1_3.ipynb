Task 1: AI-Generated Logic Without Modularization (Fibonacci Sequence
Without Functions)

```python
n = int(input("Enter number of terms: "))

a = 0
b = 1

if n <= 0:
    print("Please enter a positive number")
elif n == 1:
    print(a)
else:
    print(a, b, end=" ")
    for i in range(2, n):
        c = a + b
        print(c, end=" ")
        a = b
        b = c
```

```python
n = int(input("Enter number of terms: "))

x, y = 0, 1
if n <= 0:
    print("Please enter a positive number")
else:
    for i in range(n):
        print(x, end=" ")
        x, y = y, x + y
```

Task 3: Modular Design Using AI Assistance (Fibonacci Using Functions)

```python
def fibonacci(n):
    """Generate Fibonacci sequence up to n terms"""
    sequence = []
    a, b = 0, 1

    for i in range(n):
        sequence.append(a)
        a, b = b, a + b

    return sequence


n = int(input("Enter number of terms: "))
print(fibonacci(n))
```

Task 5: AI-Generated Iterative vs Recursive Fibonacci Approaches (Different
Algorithmic Approaches for Fibonacci Series)

Iterative Fibonacci Implementation

```python
def fibonacci_iterative(n):
    a, b = 0, 1
    for i in range(n):
        print(a, end=" ")
        a, b = b, a + b
```

Recursive Fibonacci Implementation

```python
def fibonacci_recursive(n):
    if n <= 1:
        return n
    return fibonacci_recursive(n - 1) + fibonacci_recursive(n - 2)
```
