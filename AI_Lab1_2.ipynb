# Task 1: AI-Generated Logic Without Modularization (Factorial without
Functions)

```python
# Program to calculate factorial without using functions
n = int(input("Enter a number: "))
result = 1
for i in range(1, n + 1):
    result = result * i
print("Factorial of", n, "is:", result)
```

# Task 3: Modular Design Using AI Assistance (Factorial with Functions)

```python
def calculate_factorial(number):
    factorial = 1
    for i in range(1, number + 1):
        factorial *= i
    return factorial
n = int(input("Enter a number: "))
result = calculate_factorial(n)
print("Factorial of", n, "is:", result)
```

# Task 5: AI-Generated Iterative vs Recursive Thinking

Iterative Version (AI-Generated)

```python
def factorial_iterative(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result
```

Recursive Version (AI-Generated)

```python
def factorial_recursive(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial_recursive(n - 1)
```
